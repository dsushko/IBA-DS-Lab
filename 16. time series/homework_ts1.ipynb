```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df = pd.read_excel(
    io='monthly-milk-production.xlsx',
    index_col=0, 
    parse_dates=[0],
    names=['date', 'milk']
)
```

```python
def print_ts(df, begin_idx=0, period=1, val_label=None, color='blue'):
    x = df[begin_idx::period].index
    if val_label is None:
        y = df[df.columns[0]][begin_idx::period]
    else:
        y = df[val_label][begin_idx::period]
    plt.plot(x, y, color=color)
    ax.set_xticklabels([])
    pass
```

```python
plt.figure(figsize=(15, 8))
ax = plt.subplot(211)
print_ts(df)
ax = plt.subplot(212)
sm.graphics.tsa.plot_acf(df.milk.values.squeeze(), lags=64, ax=ax)
plt.show()
```

#### Let's conduct some analysis about data we were presented:
1) We have obvious seasonality with a period of 12 months

2) There are neither outliers nor missing values

3) Dispersion is stable and needs no stabilising

4) The beginning of the series shouldn't be dropped

5) This series is not stationary

We have stable dispersion, therefore we won't use Box-Cox transformation.

Then we begin from finding a proper way to differentiate series: we have to find d, D parameters for ARIMA model.

```python
sm.tsa.seasonal_decompose(df.milk, period=12).plot()
plt.show()
```

Let's use Dickey-Fuller criteria to prove that this series is not stationary

```python
print('Dickey-Fuller criteria\'s p=%f' % sm.tsa.stattools.adfuller(df.milk[12:])[1])
```

```python
series_milk_period = 12
```

```python
df['milk_diff_period'] = df.milk - df.milk.shift(series_milk_period)
```

```python
plt.figure(figsize=(12, 6))
print_ts(df, begin_idx=series_milk_period, val_label='milk_diff_period')
```

```python
print('Dickey-Fuller\'s criteria\'s p=%f' % sm.tsa.stattools.adfuller(df.milk_diff_period[12:])[1])
```

```python
df['milk_diff'] = df.milk - df.milk.shift(1)
```

```python
plt.figure(figsize=(12, 6))
print_ts(df, begin_idx=1, val_label='milk_diff')
```

```python
print('Dickey-Fuller\'s criteria\'s p=%f' % sm.tsa.stattools.adfuller(df.milk_diff[12:])[1])
```

```python
df['milk_diff_and_period_diff'] = df.milk_diff - df.milk_diff.shift(series_milk_period)
```

```python
plt.figure(figsize=(12, 6))
print_ts(df, begin_idx=1, val_label='milk_diff_and_period_diff')
```

```python
print('Dickey-Fuller\'s criteria\'s p=%f' % sm.tsa.stattools.adfuller(df.milk_diff_and_period_diff[13:])[1])
```

At last we have totally rejected hypothesis that series is stationary, according to Dickey-Fuller's criteria.

d and D ARIMA parameters are:

d = 1

D = 1

```python
ps = range(5)
qs = range(5)
Ps = range(2)
Qs = range(2)
d = 1
D = 1
```

```python
parameters = product(ps, qs, Ps, Qs)
parameters_list = list(parameters)
```

```python
%%time
results = []
best_aic = float("inf")
warnings.filterwarnings('ignore')

for param in parameters_list:
    try:
        model=sm.tsa.statespace.SARIMAX(df.milk, order=(param[0], d, param[1]), 
                                        seasonal_order=(param[2], D, param[3], 12)).fit(disp=True)
    except ValueError:
        print('wrong parameters:', param)
        continue
    aic = model.aic

    if aic < best_aic:
        best_model = model
        best_aic = aic
        best_param = param
    results.append([param, model.aic])
    
warnings.filterwarnings('default')
```

```python
result_table = pd.DataFrame(results, columns=['parameters', 'aic'])
result_table.sort_values(by='aic', ascending=True).head()
```

```python
df['model'] = best_model.fittedvalues
```

```python
plt.figure(figsize=(15,7))
df.milk[13:].plot()
df.model[13:].plot(color='r')
plt.show()
```

```python
start = '1976-01-01'
end = '1977-12-01'

pred_dates = pd.date_range(start=start, end=end, freq='MS')
pred_vals = best_model.predict(start=start, end=end)

pred = pd.DataFrame(index=pred_dates, columns=['model'])

pred['model'] = pred_vals

df_pred = pd.concat([df, pred], axis=0)
```

Now let's check if our residuals are shifted. We are going to use Student's t-criteria:

```python
stats.ttest_1samp(best_model.resid[13:], 0)[1]
```

```python
residual_mean = np.mean(best_model.resid[13:])
residual_mean
```

```python
stats.ttest_1samp(best_model.resid[13:] - residual_mean, 0)[1]
```

According to Student's criteria we can assume that residuals are not shifted and our ARIMA model we've just built is nice as it doesn't require no explicit shift.
