**Задание 4.** Реализуйте  функцию  градиентного  спуска  для  выбора  параметров модели.  Постройте  полученную  модель  (функцию)  совместно  с графиком из пункта 2.

```python
def grad_descent_lin(X, y, step=0.01, epochs=100):
    """
    Gradient descending, that works for 2-dimensional
    linear regression.
    
     Parameters
    ----------
    X: data frame,
        feature that influences on target.
    y: Series,
        target feature.
    step: float,
        learning rate, defines step on each iteration.
    epochs: int,
        amount of steps.
        
    Returns coefficent a at x and b for
    equation y = ax + b
    
    """
    curr_a = random.random()
    curr_b = 0

    for ep in range(0, epochs):
        y_pred = [lin_f(curr_a, curr_b, x) for x in X]
        
        error = y - y_pred
        a_ddx = -2 * (X * error).mean()
        b_ddx = -2 * error.mean()

        curr_a -= step*a_ddx
        curr_b -= step*b_ddx
    return curr_a, curr_b
```

```python
def lin_f(a, b, x):
    """
    Returns value of ax + b function in point x
    a, b, x are real numbers
    """
    return a*x + b
```

```python
epochs_list = [5, 100, 250, 600, 1000, 5000]

fig, ax = plt.subplots(figsize=(15, 8))
DrawIncomeVsPopulation()

X = df['Population']

for ep in epochs_list:
    a, b = grad_descent_lin(X, y, epochs=ep)
    sns.lineplot(x=X, y=[lin_f(a, b, x) for x in X])
```

**Задание 5.*** Постройте  трехмерный  график  зависимости  функции  потерь  от параметров  модели  ($θ_0$ и  $θ_1$)  как  в  виде  поверхности,  так  и  в  виде изолиний (contour plot).

```python
xlim1 = -0.5
xlim2 = 2

ylim1 = -2
ylim2 = 2

ticks = 10

theta_0 = [x/ticks for x in range(int(xlim1*ticks),int(xlim2*ticks))]
theta_1 = [y/ticks for y in range(int(ylim1*ticks),int(ylim2*ticks))]

points_3d = []

X = df[['Population']]

for t0 in theta_0:
    for t1 in theta_1:
        population_vs_rest_income_reg = LinearRegression()
        population_vs_rest_income_reg.fit(X, [lin_f(t0, t1, x) for x in X['Population']])
        y_pred = population_vs_rest_income_reg.predict(X)
        points_3d.append((t0, t1, MSELoss(y_pred, y)))
        

fig = pylab.figure()
axes = Axes3D(fig)

ax = plt.axes(projection='3d')
ax.plot_trisurf([points_3d[i][0] for i in range(len(points_3d))], 
                [points_3d[i][1] for i in range(len(points_3d))],
                [points_3d[i][2] for i in range(len(points_3d))],
               )

ax.set_xlabel('$θ_0$')
ax.set_ylabel('$θ_1$')
ax.set_zlabel('MSE Loss')

ax.set_xlim(xlim1, xlim2)
ax.set_ylim(ylim1, ylim2)
ax.set_zlim(0, 75)

pylab.show()
```

```python
# more mindful api usage: np.linspace, plotly - contour plot

theta_0 = np.linspace(-10,10,100)
theta_1 = np.linspace(-10,10,100)
loss_vals = np.zeros((len(theta_0), len(theta_1)))

for i in range(len(theta_0)):
    for j in range(len(theta_1)):
        population_vs_rest_income_reg = LinearRegression()
        population_vs_rest_income_reg.fit(X, [lin_f(theta_0[i], theta_1[j], x) for x in X['Population']])
        y_pred = population_vs_rest_income_reg.predict(X)
        loss_vals[i][j] = MSELoss(y_pred, y)
```

**Задание 7.** Произведите  нормализацию  признаков.  Повлияло  ли  это  на  скорость сходимости градиентного спуска? Ответ дайте в виде графика.

```python
def lin_f_n(a, b, x):
    return np.dot(x[1], a) + b
```

```python
MSE_Loss_iterations_data

def grad_descent_lin_n(X, y, step=0.01, epochs=100):
    """
    Gradient descending, that works for n-dimensional
    linear regression.
    
     Parameters
    ----------
    X: data frame,
        features that influence on target.
    y: Series,
        target feature.
    step: float,
        learning rate, defines step on each iteration.
    epochs: int,
        amount of steps.
        
    Returns coefficents at x_i (a_i) as a collection and B for
    matrix equation y = AX + B,
    where 
    A = (a_1,a_2, ..., a_n),
    X = (x_1,x_2, ..., x_n)
    B is real number
    """
    global MSE_Loss_iterations_data
    MSE_Loss_iterations_data  = [0]*epochs
    curr_a = [0]*X.shape[1]
    curr_b = 0
    a_ddx = [0]*X.shape[1]
    for ep in range(0, epochs):
        y_pred = [lin_f_n(curr_a, curr_b, row) for row in X.iterrows()]    
        error = y - y_pred
        
        for col in X:
            a_ddx[X.columns.get_loc(col)] = -2 * (X[col] * error).mean()
        b_ddx = -2 * error.mean()
        
        for i in range(len(curr_a)):
            curr_a[i] -= step * a_ddx[i]
        curr_b -= step * b_ddx
        MSE_Loss_iterations_data[ep] = (error**2).mean()
        
    return curr_a, curr_b
```

```python
ax = plt.axes(projection='3d')

epochs_list = [50,100, 250, 600]
colors = {50 : 'blue', 100 : 'yellow', 250 : 'green', 600 : 'red'}
X = df[['HouseSquare', 'RoomsQuantity']]
y = df['HousePrice']
for ep in epochs_list:
    a, b = grad_descent_lin_n(X, y, epochs=ep)
    ax.scatter3D(X['HouseSquare'], X['RoomsQuantity'], [lin_f_n(a, b, row) for row in X.iterrows()], color=colors[ep])   

# Warning actually means that values became too big, and
# gradient descendeing turned out to be ascending =)
# therefore it diverges, values approach infinity
```

**Задание 8.** Реализуйте  функции  потерь  MSE  и  градиентного  спуска  для  случая многомерной линейной регрессии с использованием векторизации.

```python
# as all those functions already use vectorization
# I'm going to create 'bad' functions

def MSELoss_rough(y_pred, y_true):
    """
    MSE Loss function, but with no vectorization.
    Returns mean squared error of prediction.
    """
    diffs_sum = 0
    for i in range(len(y_pred)):
        diffs_sum += (y_true[i] - y_pred[i]) ** 2
    res = diffs_sum / len(y_true)
    return res

def grad_descent_lin_n_rough(X, y, step=0.01, epochs=100):
    """
    Gradient descending, that works for n-dimensional
    linear regression, but without vectorization.
    
    Parameters
    ----------
    X: data frame,
        features that influence on target.
    y: Series,
        target feature.
    step: float,
        learning rate, defines step on each iteration.
    epochs: int,
        amount of steps.
    
    Returns coefficents at x_i (a_i) as a collection and B for
    matrix equation y = AX + B,
    where 
    A = (a_1,a_2, ..., a_n),
    X = (x_1,x_2, ..., x_n)
    B is real number
    """
    curr_a = [0]*X.shape[1]
    curr_b = 0
    a_ddx = [0]*X.shape[1]
    for ep in range(0, epochs):
        y_pred = [lin_f_n(curr_a, curr_b, row) for row in X.iterrows()]    
        error = [y[i] - y_pred[i] for i in range(len(y))]
        
        for col in X:
            a_ddx[X.columns.get_loc(col)] = -2 * sum([X[col][i]*error[i] for i in range(len(error))])/len(error)
            
        b_ddx = -2 * np.array(error).mean()
        
        for i in range(len(curr_a)):
            curr_a[i] -= step * a_ddx[i]
        curr_b -= step * b_ddx
        
    return curr_a, curr_b
```
